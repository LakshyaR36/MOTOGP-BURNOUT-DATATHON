```python
drop_cols = ['Unique ID', 'Rider_ID', 'rider_name', 'team_name', 'bike_name']
```

```python
df = df.drop(columns=drop_cols)
```

```python
df
```

```python
X = df.drop('Lap_Time_Seconds' , axis=1)
Y = df['Lap_Time_Seconds']
```

```python
num_feat = X.select_dtypes(include=['Int64' , 'Float64']).columns.to_list()
cat_feat = X.select_dtypes(include='object').columns.to_list()
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size=0.3 , random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler , MinMaxScaler , OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , r2_score , mean_squared_error
```

```python
model = LinearRegression()
Transformers = ColumnTransformer(
    transformers=[
        ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore'), cat_feat),
        ('scaler', StandardScaler(), num_feat)
    ],
    remainder='passthrough'  
)

pipe = Pipeline([
    ('Transformers', Transformers),  
    ('model', model)    
])
```

```python
pipe
```

```python
pipe.fit(X_train , Y_train)
```

```python
Y_pred = pipe.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import mean_absolute_error , r2_score , mean_squared_error
mae = mean_absolute_error(Y_test, Y_pred)
rmse = mean_squared_error(Y_test, Y_pred, squared=False)
r2 = r2_score(Y_test, Y_pred)
print(mae)
print(rmse)
print(r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model1 = RandomForestRegressor(
    n_estimators=150,          
    max_depth=18,                  
    min_samples_leaf=2,        
    max_features='log2',                   
    n_jobs=-1,     
    bootstrap=True,            
    random_state=42
)
```

```python
pipe = Pipeline([
    ('Transformers', Transformers),  
    ('model', model1)    
])
```

```python
pipe
```

```python
pipe.fit(X_train, Y_train)
```

```python
Y_pred_RF = pipe.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error , r2_score , mean_squared_error
mae = mean_absolute_error(Y_test , Y_pred_RF)
print(mae)
rsme = mean_squared_error(Y_test , Y_pred_RF , squared=False)
print(rsme)
r2 = r2_score(Y_test, Y_pred_RF)
print(r2)
```
